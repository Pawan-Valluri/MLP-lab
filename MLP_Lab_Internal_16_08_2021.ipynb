### Write a Program to implement Lists and Dictionary

#### List

```python
lst = list(range(11))
lst
```

```python
print("Length of list :",len(lst))
```

```python
print("element at Index 2 is :",lst[2])
```

```python
print("lst[2:-2] =",lst[2:-2])
```

```python
lst.reverse()
print("list reversed :",lst)
```

```python
lst.remove(1)
print("list after removing element 1",lst)
```

```python
print("no of elements with value 1 :",lst.count(1))
```

```python
print("Popping an element :",lst.pop())
```
